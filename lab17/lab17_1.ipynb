## Навчання без вчителя. Кластеризація

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
X,y = make_blobs(random_state=1)
kmeans = KMeans(n_clusters=10, n_init=42).fit(X)
X0 = X[:,0]
X1=X[:,1]

plt.scatter(x=X0, y=X1,c=kmeans.labels_)
plt.show()
plt.scatter(X0, X1, c = y)
plt.show()
```
