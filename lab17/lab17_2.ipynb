## Навчання без вчителя. Кластеризація

```python
from pandas import read_csv
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
names=['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = read_csv(url, names=names)
X = dataset.iloc[:,:4].to_numpy()
y = dataset['class'].values
```

```python
kmeans = KMeans(n_clusters=3, n_init=42).fit(X)
```

```python
X0 = X[:,0]
X1 = X[:,1]
X2 = X[:,2]
X3=X[:,3]

plt.rcParams["figure.figsize"] = (12, 4) # (w, h)

plt.subplot(1, 3, 1)
plt.scatter(x=X0, y=X1,c=kmeans.labels_)

plt.subplot(1, 3, 2)
plt.scatter(x=X0, y=X2,c=kmeans.labels_)

plt.subplot(1, 3, 3)
plt.scatter(x=X0, y=X3,c=kmeans.labels_)


plt.tight_layout()
plt.show()
```

```python
encoder = LabelEncoder()
labels = encoder.fit_transform(y)


plt.rcParams["figure.figsize"] = (12, 4) # (w, h)

plt.subplot(1, 3, 1)
plt.scatter(x=X0, y=X1,c=labels)

plt.subplot(1, 3, 2)
plt.scatter(x=X0, y=X2,c=labels)

plt.subplot(1, 3, 3)
plt.scatter(x=X0, y=X3,c=labels)


plt.tight_layout()
plt.show()
```
